```python
Datadirectory = "/content/drive/MyDrive/CK+48"
Classes = ["0", "1", "2", "3", "4", "5", "6"]
```

```python
Data = []
img_size = 224
datagen = ImageDataGenerator(rotation_range=30)

def create_Data():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category)
        count = 0
        for img in os.listdir(path):
            
            try:
                img_array = cv2.imread(os.path.join(path, img))
                new_array = cv2.resize(img_array, (img_size, img_size))
                Data.append([new_array, class_num])

                #print("hello")
                try:
                  samples = expand_dims(img_array, 0)
                  it = datagen.flow(samples, batch_size=1)

                #print(it)
                except Exception as e:
                  print(e)
                for i in range(3):
                  batch = it.next()
                  aug_image = batch[0].astype('uint8')
                  aug_arr = cv2.resize(aug_image, (img_size, img_size))
                  # pyplot.imshow(aug_image)
                  # pyplot.show()
                  # # print(aug_arr)
                  # # plt.imshow(aug_arr)
                  # # print("hellloo")
                  # break
                  Data.append([aug_arr, class_num])
            except Exception as e:
                pass
```

```python
create_Data()
```

```python
import random

random.shuffle(Data)

X = []
y = []

for features, label in Data:
    X.append(features)
    y.append(label)
    
X = np.array(X).reshape(-1, img_size, img_size, 3)
X = X/255.0
Y = np.array(y)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

del X
del Y
del Data
del y
```
